```
#@title Run Multiple Experiments { display-mode: "form" }
import random
from  matplotlib import pyplot as plt


def experiment():
    
    coins = 256
    trial = 1
    
    countHeads = []
    countTails = []
    countCoins = []
    trials = []

    while coins >= 1:
    
        heads = 0
        tails = 0
    
        for flip in range(coins):
            flip = random.randint(0, 1)
            if flip == 0:
                heads += 1
            else:
                tails += 1
        
        countHeads.append(heads)
        countTails.append(tails)
        countCoins.append(coins)
        trials.append(trial)
    
        trial += 1
        coins = coins - tails

    plt.style.use('fivethirtyeight')
    plt.plot(trials, countCoins)
    plt.xlabel('number of trials')
    plt.ylabel('number of coins')
    plt.show()

EXPERIMENTS = int(input("How many experiments would you like to run? "))

for e in range(EXPERIMENTS):
    print(f"Experiment #{e+1}") 
    experiment()

```

```
#@title Run s Single Experiment { display-mode: "form" }
#! python3
import random
from  matplotlib import pyplot as plt

coins = 256
trial = 1

countHeads = []
countTails = []
countCoins = []
trials = []

while coins >= 1:

  heads = 0
  tails = 0

  for flip in range(coins):
    flip = random.randint(0, 1)
    if flip == 0:
      heads += 1
    else:
      tails += 1

  countHeads.append(heads)
  countTails.append(tails)
  countCoins.append(coins)
  trials.append(trial)


  print("Trial: " + str(trial))
  print("Total coins flipped: " + str(coins))
  print("Number of heads: " + str(heads))
  print("Number of tails: " + str(tails))
  print("")

  trial += 1
  coins = coins - tails

plt.style.use('fivethirtyeight')
plt.plot(trials, countCoins)
plt.xlabel('number of trials')
plt.ylabel('number of coins')
plt.show()
```
